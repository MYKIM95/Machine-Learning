```python
import pandas as pd
import matplotlib.pyplot as plt  # 그래프 그리는 라이브러리
from sklearn.model_selection import train_test_split
from sklearn import metrics  # 평가를 위한 라이브러리
from sklearn.tree import DecisionTreeClassifier
```

### 데이터 불러오기

```python
data = pd.read_csv('data/mushroom.csv')
```

```python
data.head()
```

### 목표
- 버섯의 특징을 활용해 독버섯/식용버섯 분류

```python
data.shape
```

```python
data.info()
```

```python
X = data.iloc[:,1:]
y = data.iloc[:,0]
```

```python
data.describe()
```

```python
y.value_counts()  # 중복제거된 데이터의 수를 알려줌
```

### 원 핫 인코딩

```python
X_one_hot = pd.get_dummies(X)  # 알아서 원핫 인코딩 형식으로 바꿔줌
X_one_hot
```

### 라벨 인코딩

```python
X['habitat'].unique()
```

```python
# 각각의 값에 맞게 임의로 숫자 지정
habitat_dic = {   
    'u':1,
    'g':2,
    'm':3,
    'd':4,
    'p':5,
    'w':6,
    'l':7
}
```

```python
X['habitat'].map(habitat_dic)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot , y , test_size = 0.3, random_state = 5)
```

```python
tree_model = DecisionTreeClassifier()
```

```python
tree_model.fit(X_train, y_train)
```

```python
pre = tree_model.predict(X_test)
```

```python
metrics.accuracy_score(pre, y_test)
```
